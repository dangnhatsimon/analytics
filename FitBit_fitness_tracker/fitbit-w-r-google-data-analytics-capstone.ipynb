# Import Library and Dataset

```r
#Import libraries
library (tidyverse)
library (dplyr)
library (ggplot2)
```

```r
#Import datasets

activity <- read.csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
calorie <- read.csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyCalories_merged.csv")
intensity <- read.csv ("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailyIntensities_merged.csv")
step <- read.csv ("../input/fitbit/Fitabase Data 4.12.16-5.12.16/dailySteps_merged.csv")
sleep <- read.csv ("../input/fitbit/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv")
weight_log <- read.csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/weightLogInfo_merged.csv")
```

The next step is to look at the brief statistical datasets. Because the "activity" dataset is a merge of several datasets. A statistical brief is carried out on the datasets that make up the "activity" dataset, namely calorie, intensity, step, and sleep.

```r
#calorie
calorie %>%
   select(Calories) %>%
   summary()
#intensity
intensity %>%
   select(VeryActiveMinutes, FairlyActiveMinutes, LightlyActiveMinutes, SedentaryMinutes) %>%
   summary()
#step
step %>%
   select(StepTotal) %>%
   summary()
#sleep
 sleep %>%
   select(TotalSleepRecords, TotalMinutesAsleep, TotalTimeInBed) %>%
   summary()
```

**Findings from statistics summary**
* Calorie

According to [healthline.com](https://www.healthline.com/health/fitness-exercise/how-many-calories-do-i-burn-a-day#calorie-calculator), the **total calories must be burned to depend on each individual's fitness goal**. To maintain weight, the number of calories that must be burned is equivalent to the calories that come in. With this data, we don't know whether the calories burned are enough or not.

* Intensity

[The CDC](https://www.cdc.gov/physicalactivity/basics/adults/index.htm) **recommends 30 minutes** of moderate-intensity activity per day. The data shows that the participants have had enough in their activities, which is **34.72 minutes** (Very Active + Fairly Active).

The average **sedentary participants were 991.2 minutes** or 16.52 hours. More than half of the **participants spend their time with no physical activity!**

* Step

Participants' average daily steps are 7638. The CDC recommends adults take up to 8000 steps per day. **It takes at least a 5% daily step increase to reach [the CDC's recommendations](https://www.cdc.gov/media/releases/2020/p0324-daily-step-count.html).**

* Sleep

The average participant **sleeps for 7 hours a day.**


# Process the data

```r
#Total Steps vs. Calories correlation
cor(activity$TotalSteps, activity$Calories)
#Total Steps vs. Calories visualization
ggplot (data = activity, aes(x = TotalSteps, y = Calories)) + geom_point() + labs (title = 'Total Steps vs. Calories') + geom_smooth(method = 'loess')
```


**1. Calorie**
* The level of correlation coefficient shows the number 0.59. Based on this data, **the number of steps taken has a moderate correlation to the calories burned**.
* Fitbit can **display the calorie/step ratio** in the application so that users can monitor how many calories are burned based on the number of steps only.

```r
#Grouping participant into four categories by Intensity Minutes for better visualization
participant_type <- intensity %>% 
  summarise(summary_participant_type = factor(case_when(
    SedentaryMinutes > mean(SedentaryMinutes) ~ "Passive",
    LightlyActiveMinutes > mean(LightlyActiveMinutes) ~ "Lightly Active",
    FairlyActiveMinutes > mean(FairlyActiveMinutes) ~ "Active",
    VeryActiveMinutes > mean(VeryActiveMinutes) ~ "Very Active",), 
    levels = c("Passive", "Lightly Active", "Active", "Very Active")), .group=Id) %>% 
  drop_na()
#Participant by categories percentage
participant_type %>% 
  group_by(summary_participant_type) %>% 
  summarise (total = n()) %>% 
  mutate (totals = sum(total)) %>% 
  group_by(summary_participant_type) %>% 
  summarise (TotalPercent = total / totals)
#Visualize by categories percentage
participant_type %>% 
  group_by(summary_participant_type) %>% 
  summarise (total = n()) %>% 
  mutate (totals = sum(total)) %>% 
  group_by(summary_participant_type) %>% 
  summarise (TotalPercent = total / totals) %>% 
  ggplot(aes(summary_participant_type, y = TotalPercent, fill=summary_participant_type)) + 
  geom_col() + 
  labs(title = 'Participants Distribution by Active Minutes', x= 'Category') + 
  theme(legend.position = "none") + 
  scale_y_continuous(labels = scales::percent)
```

**2. Intensity**

* If you look at the column graph, it can be seen that there are disparities between the **four categories of participants**. Most of the participants, which is around **60%**, fall into the **"Passive" category**. To further analyze why the majority of participants belong to the "Passive" category, specific data regarding the participants' activities in hours are needed.

*This analysis segment is inspired by [Julien Aranguren](https://www.kaggle.com/julenaranguren/bellabeat-case-study#PHASE-3-&-4:-Process-and-Analyse).*


```r
#Import hourly steps
steps_hourly <- read.csv("../input/fitbit/Fitabase Data 4.12.16-5.12.16/hourlySteps_merged.csv")
head(steps_hourly)
#Separate date and time
steps_hourly$ActivityHour=as.POSIXct(steps_hourly$ActivityHour, format = "%d/%m/%Y %I:%M:%S %p", tz=Sys.timezone())
steps_hourly$time <- format(steps_hourly$ActivityHour, format = "%H:%M:%S")
steps_hourly$date <- format(steps_hourly$ActivityHour, format = "%m/%d/%y")
head(steps_hourly)
#Average steps by hour
steps <- steps_hourly %>% 
  group_by(time) %>% 
  drop_na() %>% 
  summarise(mean_total_steps = mean(StepTotal))
#Visualize
ggplot(data = steps, aes(x=time, y=mean_total_steps)) + 
  geom_histogram(stat = "identity", fill = "blue") + 
  theme (axis.text.x = element_text(angle = 90)) +
  labs (title = "Average Total Steps vs. Time")
```

**3. Step**

* From the histogram, we can see that the average step of the participants is **relatively high in the morning at 9 A.M.** This is usually the time people start work, and they are likely to travel on foot. Then it **rises again at 12 P.M. - 1 P.M.** where the hours are work breaks. The average daily step **peaks at 5 P.M. - 7 P.M.** This is the time when people finish work, and it seems several participants are exercising during this period (ex. walk, run, gym session).

*This analysis segment is inspired by [Anastasiia Chebotina](https://www.kaggle.com/chebotinaa/bellabeat-case-study-with-r).*

```r
#Grouping participant into three categories by sleep time
sleep_type <- sleep %>% 
  summarise(summary_participant_sleep = factor(case_when(
    TotalMinutesAsleep < 420 ~ "Need more sleep",
    TotalMinutesAsleep >= 420 & TotalMinutesAsleep < 540 ~ "Good sleep",
    TotalMinutesAsleep > 540 ~ "Need less sleep",),
    levels = c("Need more sleep", "Good sleep", "Need less sleep")), .group =Id) %>% 
  drop_na()
#Participant by category percentage
sleep_type %>% 
  group_by(summary_participant_sleep) %>% 
  summarise (total = n()) %>% 
  mutate (totals = sum(total)) %>% 
  group_by(summary_participant_sleep) %>% 
  summarise (TotalPercent = total / totals)
#Visualize
sleep_type %>% 
  group_by(summary_participant_sleep) %>% 
  summarise (total = n()) %>% 
  mutate (totals = sum(total)) %>% 
  group_by(summary_participant_sleep) %>% 
  summarise (TotalPercent = total / totals) %>% 
  ggplot(aes(summary_participant_sleep, y= TotalPercent, fill=summary_participant_sleep)) + 
  geom_col() + 
  labs(title='Participants Distribution by Sleep Time', x = 'Category') + 
  theme(legend.position = 'none') +
  scale_y_continuous(labels = scales::percent)
```

```r
#Sleep Time vs. Time in Bed correlation
cor(sleep$TotalTimeInBed, sleep$TotalMinutesAsleep)
#Sleep Time vs. Time in Bed visualization
ggplot(data = sleep, aes(x=TotalTimeInBed, y=TotalMinutesAsleep)) + geom_point() + 
  labs(title = 'Total Time in Bed vs. Total Minutes Sleep') + geom_smooth()
```

**4. Sleep**

* According to [the CDC](https://www.cdc.gov/sleep/about_sleep/how_much_sleep.html), the recommended total hours of sleep for adults is more than 7 hours. Meanwhile, according to [Sleepfoundation.org](https://www.sleepfoundation.org/how-sleep-works/how-much-sleep-do-we-really-need#:~:text=National%20Sleep%20Foundation%20guidelines1,to%208%20hours%20per%20night.), the number of hours of sleep more than 9 hours for adults is an excessive amount. From the data visualization, we can see that **only 46% of the participants had a good number of hours of sleep**. Therefore, it is necessary to have a **notification reminder to sleep and a built-in alarm feature** to increase engagement with consumers.
* The correlation between Total Time in Bed and Total Minutes Sleep is very significant, namely r = 0.93. However, **several participants showed a high difference between Total Time in Bed and Total Minutes Sleep** when viewed further. This data can be displayed to consumers regularly so that if consumers who have a data pattern like this want to increase the number of hours of sleep, they can choose the **'Reminder go to sleep'** feature.

# Recommendation

* Based on [FitBit Fitness Tracker Data](https://www.kaggle.com/arashnic/fitbit), there is no **daily calories intake**, so we cannot determine whether the calories burned are sufficient or excessive. In addition to providing manual calories input, Fitbit can provide food or drink options along with the average calories contained in it, making it easier for users to use this feature.
* With an average Sedentary Time of 16 ½ hours, Fitbit can provide a **reminder to its users to get to work immediately**. So that the reminder does not disturb the user, Fitbit can create a **'Work Time' feature** where Fitbit will not provide a reminder that bothers the user while they are working.
* Daily steps are still below [the CDC recommendation](https://www.cdc.gov/media/releases/2020/p0324-daily-step-count.html) of 8000. Fitbit can provide the **'Goal Level' feature**. That is the daily step goals feature which is divided into several levels. Examples are three levels, Passive Level (2500 steps), Lightly Active Level (5000 steps), and Active Level (≥8000 steps). When the user can pass the Passive Level several times in a row, Fitbit will recommend the user to take 5000 steps (level up) and so on so that the feature can provoke the user to take more steps gradually.
* More than half of the participants did not have enough hours of sleep. In addition, several participants have a significant difference in Time in Bed and Time Asleep (I assume leading to Sedentary Time). Fitbit should **give the result of daily sleep hours when the user wakes up** so that if the user wants to increase/decrease the number of hours of sleep, the user can set a **reminder to go to sleep immediately or a built-in alarm** if the target number of hours of sleep has been reached.
